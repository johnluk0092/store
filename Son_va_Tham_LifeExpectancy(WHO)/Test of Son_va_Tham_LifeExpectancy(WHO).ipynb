## Nội dung vấn đề:

    Mặc dù đã có nhiều nghiên cứu về các yếu tố ảnh hưởng đến tuổi thọ khi dựa trên các biến của nhân khẩu học, thu nhập và tỷ lệ tử vong. 

    Trước đây đã không tính đến ảnh hưởng của chỉ số tiêm chủng và phát triển con người. 

    Ngoài ra, một số nghiên cứu trước đây đã được thực hiện multiple linear regression dựa trên tập dữ liệu của một năm cho tất cả các quốc gia. 

    Do đó, điều này tạo động lực để giải quyết cả hai yếu tố đã nêu trước đây bằng cách xây dựng mô hình a regression model dựa trên mô hình tác động hỗn hợp và multiple linear regression trong khi xem xét dữ liệu từ giai đoạn 2000 đến 2015 cho tất cả các quốc gia.

    Việc chủng ngừa quan trọng như Viêm gan B, Bại liệt và Bạch hầu cũng sẽ được xem xét. 

    Tóm lại, nghiên cứu này sẽ tập trung vào các yếu tố miễn dịch, yếu tố tử vong, yếu tố kinh tế, yếu tố xã hội và các yếu tố liên quan đến sức khỏe khác.

    Vì các observations trong tập dữ liệu này dựa trên các quốc gia khác nhau, nên một quốc gia sẽ dễ dàng hơn trong việc xác định yếu tố dự đoán đang góp phần làm giảm giá trị tuổi thọ.

    Điều này sẽ giúp gợi ý cho một quốc gia về lĩnh vực nào cần được coi trọng để nâng cao tuổi thọ của dân số một cách hiệu quả.

## Nội Dung Dữ liệu:

    Tập dữ liệu dựa trên kho dữ liệu của Global Health Observatory (GHO) thuộc Tổ chức Y tế Thế giới (WHO) theo dõi tình trạng sức khỏe cũng như nhiều yếu tố liên quan khác của tất cả các quốc gia. 

    Tập dữ liệu liên quan đến tuổi thọ, các yếu tố sức khỏe của 193 quốc gia đã được thu thập từ cùng một trang web kho dữ liệu của WHO và dữ liệu kinh tế  được thu thập từ trang web của United Nation. 

    Tập dữ liệu có dữ liệu từ năm 2000-2015 của 193 quốc gia 

    Khi kiểm tra trực quan ban đầu của dữ liệu cho thấy một số giá trị bị thiếu. Vì bộ dữ liệu được lấy từ WHO. Dữ liệu bị thiếu đã được xử lý trong phần mềm R bằng lệnh Missmap. 

    Kết quả chỉ ra rằng hầu hết các dữ liệu bị thiếu là về dân số, Viêm gan B và GDP. Dữ liệu bị thiếu là từ các quốc gia ít được biết đến như Vanuatu, Tonga, Togo, Cabo Verde, v.v. Việc tìm kiếm tất cả dữ liệu cho các quốc gia này rất khó khăn và do đó đã loại trừ các quốc gia này khỏi tập dữ liệu mô hình cuối cùng.

**Mô tả dữ liệu:**

* Country (Nominal) - tên của các nước trên thế giới, gồm 193 nước
* Year (Ordinal) - năm dương lịch của các chỉ số (dao động từ 2000 đến 2015)
* Status (Nominal) - một quốc gia được coi là 'Đang phát triển' hay 'Đã phát triển' theo các tiêu chuẩn của WHO
* Life expectancy (Ratio) - tuổi thọ trung bình của người dân tính theo năm đối với một quốc gia và năm cụ thể
* Adult Mortality (Ratio) - tỷ lệ tử vong của người trưởng thành trên 1000 dân số (tức là số người chết từ 15 đến 60 tuổi trên 1000 dân số), nếu tỉ lệ là 263 nghĩa là có 263 người chết trong số 1000 người có độ tuổi từ 15 đến 60. một cách khác để giải thích điều này là xác suất một cá nhân chết trong khoảng từ 15 đến 60 là 26,3%
* infant deaths (Ratio) - số trẻ sơ sinh tử vong trên 1000 dân số
* Alcohol (Ratio) - Tỷ lệ tiêu thụ rượu của một quốc gia được đo bằng lít/người
* percentage expenditure(Ratio) - chi tiêu cho y tế tính theo tỷ lệ phần trăm của Tổng sản phẩm quốc nội (gdp)
* Hepatitis B (Ratio) - số trẻ 1 tuổi được tiêm ngừa viêm gan B trên tất cả các trẻ 1 tuổi trong dân số
* Measles (Ratio) - số ca mắc bệnh Sởi được báo cáo trên 1000 dân số
* BMI (Interval/Ordinal) - Chỉ số khối cơ thể trung bình (BMI) của tổng dân số của một quốc gia
* under-five deaths (Ratio) - số người dưới 5 tuổi tử vong trên 1000 dân số
* Polio (Ratio) - số trẻ 1 tuổi được tiêm ngừa vắc xin bại liệt trên tổng số trẻ 1 tuổi trong dân số
* Total expenditure (Ratio) - chi tiêu chính phủ cho y tế tính theo tỷ lệ phần trăm tổng chi tiêu chính phủ
* Diphtheria (Ratio) - Tỷ lệ tiêm chủng uốn ván và ho gà (DTP3) của trẻ 1 tuổi
*  HIV/AIDS (Ratio) - số trẻ dưới 5 tuổi chết do HIV / AIDS trên 1000 ca sinh.
* GDP (Ratio) - Tổng sản phẩm quốc nội bình quân đầu người
* Population (Ratio) - dân số của một quốc gia
*  thinness  1-19 years (Ratio) - tỷ lệ gầy của những người từ 10-19 tuổi (Lưu ý: biến nên được đổi tên thành thinness_10-19_years để thể hiện chính xác hơn ý nghĩa và nội dung biến)
*  thinness 5-9 years (Ratio) - tỷ lệ gầy của trẻ từ 5 đến 9 tuổi
* Income composition of resources (Ratio) - Chỉ số Phát triển Con người xét về cơ cấu thu nhập của các nguồn lực (chỉ số từ 0 đến 1)
* Schooling (Ratio) - số năm đi học trung bình của dân số

```python
gdp = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)')
gdp1 = gdp[3]
gdp2 = gdp[4]
```

```python
gdp3 = gdp1.merge(gdp2, how='inner', on='Country (or dependent territory)')
gdp3.rename(columns = {'Country (or dependent territory)':'Country'}, inplace = True)
gdp3.head(3)
```

```python
gdp4 = gdp3.query('Country == ["Bahamas", "Bolivia", "Republic of the Congo", "Czech Republic","Democratic Republic of the Congo", "Egypt", "Eritrea", "Gambia","Iran", "Iraq", "Kyrgyzstan","Laos", "Libya","Federated States of Micronesia", "Monaco", "Niue","Papua New Guinea", "South Korea", "Moldova","Saint Kitts and Nevis", "Saint Lucia","Saint Vincent and the Grenadines", "San Marino","São Tomé and Príncipe", "Slovakia","Syria","North Macedonia","Tanzania", "United States","Venezuela", "Vietnam", "Yemen", "United Kingdom"]')
gdp4 = pd.DataFrame(gdp4.iloc[:, :-4]).reset_index()
gdp4 = gdp4.iloc[:, 1:]

gdp4.rename(columns = {'2000':2000, '2001': 2001, '2002': 2002, '2003': 2003, '2004': 2004, '2005': 2005, '2006': 2006, '2007': 2007, '2008': 2008, '2009':2009, '2010': 2010, '2011': 2011, '2012': 2012, '2013': 2013, '2014': 2014, '2015': 2015}, inplace = True)
gdp4.head(3)
```
